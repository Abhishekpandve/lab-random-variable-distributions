## Lab | 3 | Check transformation

```python
# Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm

# Assuming 'data' is your DataFrame
columns = data.columns

for column in columns:
    # Plot the distribution
    sns.displot(data[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.show()

    # Q-Q plot for normality
    sm.qqplot(data[column], line='s')
    plt.title(f'Q-Q Plot for {column}')
    plt.show()

    # Anderson-Darling test for normality
    result = stats.anderson(data[column])
    print(f"Anderson-Darling test result for {column}: {result}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import expon
import statsmodels.api as sm

# Assuming 'data' is your DataFrame
column_name = 'Customer_Lifetime_Value'

# Plot the distribution
sns.displot(data[column_name], kde=True)
plt.title(f'Distribution of {column_name}')
plt.show()

# Q-Q plot for exponentiality
sm.qqplot(data[column_name], dist=expon, line='s')
plt.title(f'Q-Q Plot for {column_name} against Exponential Distribution')
plt.show()

# Perform the Kolmogorov-Smirnov test for exponentiality
ks_statistic, ks_p_value = stats.kstest(data[column_name], 'expon')
print(f"Kolmogorov-Smirnov test result for {column_name}: Statistic = {ks_statistic}, P-value = {ks_p_value}")
```
